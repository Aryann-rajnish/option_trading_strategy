```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("20-03-2017-TO-19-03-2019HDFCEQN.csv",index_col=2, parse_dates = True)
```

```python
data.head()
```

```python
type(data)
```

```python
data = data.drop(columns=["Symbol","Series"])
```

```python
data.head()
```

```python
data = data.rename(index=str,columns={"Prev Close":"PClose","Open Price":"Open","High Price":"High","Low Price":"Low","Last Price":"Last","Close Price":"Close","Average Price":"Average"})
```

```python
data.head(3)
```

```python
data['high'] = data.Close.shift(1).rolling(window = 5).max()
```

```python
data.head(10)
```

```python
data['low'] = data.Close.shift(1).rolling(window = 5).min()
```

```python
data['avg'] = data.Close.shift(1).rolling(window = 5).mean()
```

```python
data.head(10)
```

```python
data['long_entry'] = data.Close > data.high
data['short_entry'] = data.Close < data.low
```

```python
data.head(15)
```

```python
data['long_exit'] = data.Close < data.avg
data['short_exit'] = data.Close > data.avg
```

```python
data.head(15)
```

```python
data['positions_long']= np.nan
data.loc[data.long_entry, 'positions_long'] = 1
data.loc[data.long_exit, 'positions_long'] = 0
data.head(15)
```

```python
data['positions_short']= np.nan
data.loc[data.short_entry, 'positions_short'] = -1
data.loc[data.short_exit, 'positions_short'] = 0
```

```python
data
```

```python
data["signal"] = data.positions_long + data.positions_short
data = data.fillna(method = 'ffill')
```

```python
data
```

```python
daily_log_returns = np.log(data.Close/data.Close.shift(1))
daily_log_returns = daily_log_returns * data.signal.shift(1)
daily_log_returns
```

```python
daily_log_returns.hist(bins = 50)
plt.show()
```

```python
daily_log_returns.cumsum()
```

```python
daily_log_returns.cumsum().plot()
plt.show()
```
